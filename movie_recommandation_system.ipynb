```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
#ps is object
```

```python
#now we are creating helper function for replacing each word of the string
def stem(text):
    y = []
    
    for i in text.split(): #for splitting the string so we can use each word 
        y.append(ps.stem(i))
        
    return " ".join(y) #for joinnig back to y to form string
```

```python
#now we will check for random string using steam 
stem("loved loving loves loving")
```

```python
#now we will apply to tag 
new_df['tag']=new_df['tag'].apply(stem)
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
#here cv is an object 
#max_feature is an hyperparameter for selecting maximum feature 
#stop_words will use for removing common laguage jointer like is this are on etc
```

```python
 vectors=cv.fit_transform(new_df['tag']).toarray()
#by default many values will be 0's and
#countvector will produce result in scipy sparsematrix formation 
#we need data in numpy array so we will convert it to numpy array using
#toarray()
#using fit_transform we are transforming them to vector form
```

```python
#now we will not calculate euclidian's distance wi will use cosine distance 
#because euclidians distance will fail in iarge dimension 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#from using cosine_similarity we will calculate cosine distance of one vectors with other all vector
#we are using sklearn library 
#we will get data in metrics formation
```

```python
simillarity=cosine_similarity(vectors)
# we are finding distance of 4806 4806 with all 4806 vectors
#we are storing diance metrics in simillarity variable
```

```python
list(enumerate(simillarity[0]))
#enumerate will create  index with distance positon so that if we order the distances their index will not be lost
```

```python
#now we will create function which will return simillar 5 movies 
#of the movie we have passed
#step :1 we will fetch index of movie
def recommend(movie):
    movie_index=new_df[new_df['title'] == movie ].index[0]
    distances = simillarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    # key=lambda x:x[1] will use for sort according to simillarity score not index because index is at 0 position and score at 1 positon
    #[1,6] is used to print top 5 nearest simillarity vector
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
